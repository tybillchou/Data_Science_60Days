# Seaborn圖樣式
有五種預設seaborn主題：darkgrid，whitegrid，dark，white，和ticks。它們各自適合不同的應用和個人喜好。默認主題是darkgrid

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

作業(1)

顯示的四個變量中, 把hue 套用 sex

x="total_bill", y="tip", hue="sex", style="time"

```python
# 載入數據集
tips = sns.load_dataset("tips")
fmri = sns.load_dataset("fmri")
```

```python
#觀看資料集有哪些變數
tips.head()
```

```python
# 通過獨立地改變每個點的色調和样式，也可以表示四個變量
'''
顯示的四個變量中, 把hue 套用 sex
x="total_bill", y="tip", hue="sex", style="time"

在此編寫
'''
sns.lmplot(x="total_bill", y="tip" ,data= tips, hue="sex" )


```

```python
sns.lmplot(x="total_bill", y="tip" ,data= tips, hue="sex", col="time" )
```

```python
#改用新的資料集
#觀看資料集有哪些變數
'''
在此編寫
'''
fmri.head()
```

```python
#使用新的數據集fmri, 再跑一次
#選用特徵值 x="timepoint", y="signal", hue="event", style="region", kind=line(表示用線條)
'''
在此編寫
'''
sns.relplot(x="timepoint", y="signal" ,data= fmri, hue="event", style="region")
```

```python
#使用新的數據集fmri, 再跑一次 sns.regplot
# x="timepoint", y="signal"
'''
在此編寫
'''
sns.regplot(x="timepoint", y="signal" ,data= fmri)
```
