```python
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt

# 新建地圖
map = Basemap(projection='ortho',lat_0=0,lon_0=0)  #Basemap有很多屬性，這里全都使用默認参數

map.fillcontinents(color='coral', lake_color='aqua')

# 畫圖
map.drawcoastlines()

# 顯示结果
plt.show()

# 存储结果
plt.savefig('test.png')
```

```python
map = Basemap(projection='mill',lat_0=0,lon_0=0)

map.drawcoastlines()
plt.show()
```

```python
map = Basemap(projection='mill', resolution='l', llcrnrlon=-180, llcrnrlat=-90, urcrnrlat=90, urcrnrlon=180)

map.drawcountries(color='darkred')
map.drawcoastlines(color='r')
map.drawmapboundary(fill_color='aqua')#海顏色
map.fillcontinents(color = 'coral')#陸地顏色
plt.show()
```
